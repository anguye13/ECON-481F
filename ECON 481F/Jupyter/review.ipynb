# Visualization

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,200)
y = np.sin(x)
z = np.cos(x)

plt.plot(x,y)
plt.plot(x,z,'-r')
```

```python
fig, ax = plt.subplots()

ax.plot(x,y,label='sin')
ax.plot(x,z,label='cos')
ax.legend()
ax.set_yticks((-1,0,1))
ax.set_title('plot')
```

```python
countries = ['CAN','MEX','USA']
populations = [36.7,129.2,325.7]
land_area = [3.85, 0.761, 3.79]

fig, ax = plt.subplots(2)
ax[0].bar(countries, populations)
ax[0].set_title('pops')

ax[1].bar(countries, land_area)
ax[1].set_title('land')

fig.tight_layout()
```

```python
N = 50

rng = np.random.default_rng()

x = rng.random(50)
y = rng.random(50)
size = rng.random(50)*50

fig, ax = plt.subplots()

ax.scatter(x,y,s=size,c=rng.random(50),alpha=.5)
```

```python
rng = np.random.default_rng()

x = np.linspace(800,1100,300)
y = rng.random(300)

fig, ax = plt.subplots()
ax.plot(x,y)
```

### Scipy

The probability that an adult man in the labor force who is unemployed is about 5%. A random sample of 10 men is taken.

Find the probability that

- all 10 men are unemployed
- no men are unemployed
- exactly 2 men are unemployed
- at least 2 men are unemployed

```python
from scipy.stats import binom

binom.pmf(10,10,.05)
```

```python
binom.pmf(0,10,.05)
```

```python
binom.pmf(2,10,.05)
```

```python
1-binom.pmf(0,10,.05)-binom.pmf(1,10,.05)
```

```python
1-binom.cdf(1,10,.05)
```

```python
np.mean(binom.rvs(10,.05,size=100000))
```

```python
means = np.zeros(10000)

for i in range(10000):
    means[i] = np.mean(binom.rvs(10,.05,size=10))

fig, ax = plt.subplots()

ax.hist(means)
plt.show()
```
