```python
#Problem2 + Problem4

gdp = pd.read_csv(url, parse_dates=True, index_col=0)
```

```python
#Problem3
names = {"GDPC1":"GDP","PCECC96":"Consumption","GPDIC1":"Investment"}
gdp.rename(columns=names,inplace=True)
```

```python
#Problem 5.1

gdp.loc['1/1/2000':'12/31/2009']
```

```python
#Problem 5.2

gdp_growth = gdp.pct_change()

names = {"GDP":"GDP Growth","Consumption":"Consumption Growth","Investment":"Investment Growth"}
gdp_growth.rename(columns=names,inplace=True)

gdp_growth
```

```python
#Problem 5.3

gdp_growth['GDP Growth'].idxmax()
```

```python
#Problem 5.4

def recession(s):
    if s.min() < 0:
        return 'recession'
    else:
        return 'normal'

df = gdp_growth.agg(recession,axis=1)
df
```
