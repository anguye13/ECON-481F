**Exercise**

Draw 10, 200, and 10000 random variables from the uniform distribution, and plot histogram

```python
ten = uniform.rvs(size=10)
two_hundred = uniform.rvs(size=200)
ten_k = uniform.rvs(size=10000)

fig, ax = plt.subplots(3, figsize=(10,10))

ax[0].set_title('10 draws')
ax[0].hist(ten)

ax[1].set_title('200 draws')
ax[1].hist(two_hundred)

ax[2].set_title('10k draws')
ax[2].hist(ten_k)
```

# Normal Distribution

```python
from scipy.stats import norm
```

**Exercise**

Draw 10, 200, and 10000 random variables from the normal distribution, and plot histogram

Draw pdf and cdf.

```python
d10 = norm.rvs(size=10)
d200 = norm.rvs(size=200)
d10k = norm.rvs(size=10000)

plot, ax = plt.subplots(3, figsize=(10,10))

ax[0].hist(d10)

ax[1].hist(d200)

ax[2].hist(d10k)
```

```python
x = np.linspace(-3,3)
pdf_x = norm.pdf(x)
cdf_x = norm.cdf(x)

plot, ax = plt.subplots(2)

ax[0].plot(x,pdf_x)
ax[1].plot(x,cdf_x)
```

# Solow Growth Model

```python
import numpy as np
import matplotlib.pyplot as plt
```

$F(K,L) = AK^{\alpha}L^{1-\alpha}$

```python
α = 1/3
```

```python
def F(A,K,L):
    return A*(K**α)*(L**(1-α))
```

```python
K = np.linspace(0,40,60)
```

```python
fig, ax = plt.subplots()
ax.plot(K, F(1,K,8))
```

```python
K1 = 8
L1 = 12
Y1 = F(1,K1,L1)
Y1
```

```python
K2 = 16
L2 = 24
Y2 = F(1,K2,L2)
Y2
```
