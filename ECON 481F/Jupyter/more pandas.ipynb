# Indexing in Pandas

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
url = "https://datascience.quantecon.org/assets/data/state_unemployment.csv"
unemp_raw = pd.read_csv(url) #parse_dates=["Date"]
```

```python
unemp_raw
```

```python
unemp_all = unemp_raw.reset_index().pivot_table(index=['Date'],columns=['state'],values='UnemploymentRate')
```

```python
unemp_all.head()
```

```python
states = [
    "Arizona", "California", "Florida", "Illinois",
    "Michigan", "New York", "Texas"
]

unemp = unemp_all[states]
```

```python
unemp.head()
```

```python
unemp = unemp.set_index(pd.to_datetime(unemp.index))
```

```python
unemp.index
```

```python
unemp.head()
```

```python
unemp.head()
```

```python
unemp.head()
```

```python
unemp.shift(0)
```

```python
unemp.resample('QS').mean()
```
